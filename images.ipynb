```python
import matplotlib.pyplot as plt

# Configuración global de tamaño de fuente
plt.rcParams.update({'font.size': 14})

# Primera gráfica: Coherencia por número de temas
num_temas = [3, 4, 5]
coherencia = [0.34, 0.35, 0.43]

plt.figure(figsize=(8, 6))
plt.plot(num_temas, coherencia, marker='o', linestyle='-', color='b')
plt.xlabel("número de temas", fontsize=14)
plt.ylabel("Coherencia", fontsize=14)
plt.title("Coherencia por número de temas", fontsize=16)
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

# Segunda gráfica: Distribución de temas en los comentarios de la categoría 'otro'
temas = ["Tema 1", "Tema 4", "Tema 5", "Tema 2", "Tema 3"]
comentarios = [19, 17, 16, 10, 9]
colores = ["yellow", "green", "teal", "navy", "purple"]

plt.figure(figsize=(8, 6))
plt.barh(temas, comentarios, color=colores)

# Agregar etiquetas de valores
for index, value in enumerate(comentarios):
    plt.text(value + 0.5, index, str(value), va='center', fontsize=14, fontweight='bold')

plt.xlabel("número de comentarios", fontsize=14)
plt.ylabel("Tema", fontsize=14)
plt.title("Distribución de temas en los comentarios de la categoría 'otro'", fontsize=16)
plt.grid(True, linestyle='--', alpha=0.6, axis='x')
plt.show()
```
